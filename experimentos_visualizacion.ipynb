```python
mujeres = df2['¿Mujeres?'].values
```

```python
mujeres_count = []
```

```python
for i in range(len(mujeres)):
    aux = mujeres[i].split(';')
    # aux = mujeres[i].
    print(aux)    
    if len(aux) == 1:
        if aux[0] == '0':
            mujeres_count.append(0)
        else:
            mujeres_count.append(1)
    else:
        mujeres_count.append(len(aux))
```

```python
institutos_count = []
```

```python
institutos = df2.Institutos.values
institutos[:-1]
```

```python
for instituto in institutos[:-1]:
    print(instituto.split(';'))
```

```python
for instituto in institutos[:-1]:
    instituto = instituto.split(';')
    institutos_count.append(instituto)
```

```python
institutos_count
```

```python
institutos = {}
```

```python
for i in range(len(institutos_count)):
    for j in range(len(institutos_count[i])):
        key = institutos_count[i][j].lstrip()
        # print(i,j,key)
        print(key)
        if key in institutos:
            institutos[key] += 1
        else:
            institutos[key] = 1
```

```python
institutos
```

```python
institutos
```

```python
inst = pd.DataFrame(institutos.items(), columns=['Instituto', 'Count'])
```

```python
inst
```

```python
sns.barplot(data=inst, x = 'Instituto', y = 'Count')
```
